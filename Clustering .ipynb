```python
#Hierarchical Clustering
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import AgglomerativeClustering
```

```python
X = np.array([[5,3],
    [10,15],
    [15,12],
    [24,10],
    [30,30],
    [85,70],
    [71,80],
    [60,78],
    [70,55],
    [80,91],])
```

```python
#Method to predict the clusters that each data point belongs to
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
cluster.fit_predict(X)
```

```python
#Mapping the clusters
plt.scatter(X[:,0],X[:,1], c=cluster.labels_, cmap='rainbow')
plt.savefig('filename.png', dpi=300)
```

```python
#K-means clustering
from sklearn.cluster import KMeans
import pandas as pd
```

```python
#Creating variables
Data = {'x': [20,33,20,27,19,34,29,31,22,35,34,67,54,57,43,50,57,59,52,65,47,49,
              48,35,33,44,45,38,43,51,46],
        'y': [77,55,57,78,65,56,73,73,57,69,75,51,32,40,47,53,36,35,58,59,50,25,20,
              14,12,20,5,29,27,8,7]
       }
```

```python
#Transforming the data into a Pandas dataFrame and then implementing the k-means algorithm
df = pd.DataFrame(Data,columns=['x','y'])

kmeans = KMeans(n_clusters=3).fit(df)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
plt.scatter(df['x'], df['y'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.savefig('filename.png', dpi=300)
```
